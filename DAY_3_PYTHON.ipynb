***GRADE INPUT SYSTEM***

```python
grade= input("Enter your grade: ")

if grade == "A":
  print("You scored above 80")

elif grade =="B":
  print("You scored between 71 and 80")

elif grade =="C":
  print("You scored between 61 and 70")

elif grade =="D":
  print("You scored between 51 and 60")

elif grade =="E":
  print("You scored between 41 and 50")

elif grade =="F":
  print("You scored less than 41")

else:
  print("Invalid input")
```

For ***or*** as long as one of the conditions is true then the entire evaluation is true

For ***and*** everything needs to be true for the final evaluation to be true

```python
a=1
b=2
c=3
print(a==b and b==c)
print(a<b or b>c)
print(a<b and b>c)
```
