```python
fn = input("First number? ")
sn = input("Second number? ")

f = float(fn) #first number
s = float(sn) #second number

print(f"The sum is {f + s}"+".")
print(f"The mutiple is {f*s}"+".")
print(f"The quotient is {f//s}"+".")
print(f"The difference is {f-s}"+".")
print(f"The quotient is {f/s}"+".")
print(f"The remainder is {f%s}"+".")
print(f"The exponential is {f**s}"+".")
```
